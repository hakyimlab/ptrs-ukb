```python
import sys
sys.path.append('../../code/')
import gcta_helper as ghelper
import numpy as np
import h5py
import pandas as pd
```

```python
list_pred_expr = []
col_order = None
genes = set()
predictor_gene_column = 'gene'
predictor_tables = [ '/vol/bmd/yanyul/UKB/gcta_regulability/pred_expr_ctimp_Muscle_Skeletal_x_Chinese.tsv', '/vol/bmd/yanyul/UKB/gcta_regulability/pred_expr_ctimp_Whole_Blood_x_Chinese.tsv' ]
ntotal = len(predictor_tables)
for i in range(ntotal):
    filename = predictor_tables[i]
    colname = predictor_gene_column  # args.predictor_gene_column
    df_pred_expr = ghelper.tsv_to_pd_df(filename, indiv_col = colname)
    # df_pred_expr = df_pred_expr.drop(columns = [colname])
    if col_order is None:
        col_order = list(df_pred_expr.columns)
        gene_idx = col_order.index(colname)
        col_order.pop(gene_idx)
        col_order.append(colname)
    else:
        df_pred_expr = df_pred_expr[col_order]
    list_pred_expr.append(df_pred_expr)
    genes = genes.union(set(df_pred_expr[colname].to_list()))
```

```python
for g in genes:
    print(g)
```

```python
piled_pred_expr = None
max_ = 2
counter = 0
for g in ['sdas']: # genes:
    if counter < max_:
        counter += 1
    else:
        break
    _mat = []
    for ele in list_pred_expr:
        _tmp = ele.loc[ele[colname] == g].to_numpy()
        _tmp = _tmp[0, :][: (_tmp.shape[1] - 1)]
        _mat.append(_tmp)
    _mat = np.array(_mat)
    _mat = ghelper.standardize_row(_mat)
    TtT = np.matmul(_mat, _mat.T) / _mat.shape[1]
    w, v = np.linalg.eigh(TtT.astype(float))
    w, v = ghelper.truncate_evd(w, v)
    
```
